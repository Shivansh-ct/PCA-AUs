```python
# Define the tick formatter function
import matplotlib.ticker as ticker

def exp_formatter(x, pos):
    """
    Return the exponential notation of x.
    """
    return '{:d}'.format(round(np.exp(x)))

# Set the x-axis tick formatter
```
